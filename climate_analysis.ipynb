```python
# some tests for datawrapper color from Chat GPT

def classify_color(anomaly):
    if anomaly <= -2.0:
        return '#08306b'      # sehr kalt – dunkelblau
    elif anomaly <= -1.0:
        return '#2171b5'      # kalt – mittelblau
    elif anomaly <= -0.5:
        return '#6baed6'      # etwas kalt – hellblau
    elif anomaly < 0.0:
        return '#c6dbef'      # leicht kalt – sehr hellblau
    elif anomaly < 0.5:
        return '#fdd0a2'      # leicht warm – sehr hellorange
    elif anomaly < 1.0:
        return '#f16913'      # etwas warm – orange
    elif anomaly < 2.0:
        return '#d94801'      # warm – dunkelorange
    else:
        return '#a63603'      # sehr warm – dunkelrot

```

```python
def classify_color_11(anomaly):
    if anomaly <= -2.0:
        return '#08306b'   # sehr kalt – sehr dunkelblau
    elif anomaly <= -1.5:
        return '#2171b5'
    elif anomaly <= -1.0:
        return '#4292c6'
    elif anomaly <= -0.5:
        return '#6baed6'
    elif anomaly < 0.0:
        return '#c6dbef'
    elif anomaly < 0.5:
        return '#fdd0a2'   # nahe am Mittelwert
    elif anomaly < 1.0:
        return '#f16913'
    elif anomaly < 1.5:
        return '#d94801'
    elif anomaly < 2.0:
        return '#a63603'
    elif anomaly < 2.5:
        return '#843c03'
    else:
        return '#662506'   # sehr warm – sehr dunkelrot
```
