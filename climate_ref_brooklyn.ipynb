```python
import requests 
import datetime
import pandas as pd
import altair as alt
import json


from dotenv import load_dotenv
import os
```

```python
df = pd.read_csv('raw/ref_brooklyn.csv', header=2)
```

```python
# rename columns

(
    df
    .rename(
        columns={
            'time': 'day', 
            'temperature_2m_mean (°C)' : 'temperature'
        }, 
        inplace=True
    )
)
```

```python
# add month to data
df['day'] = pd.to_datetime(df['day'])
df.dtypes
```

```python
df['year'] = df['day'].dt.to_period('Y')
df['month'] = df['day'].dt.month
```

```python
# calculate ref period for every month from 1961-1990

df.groupby('month')['temperature'].mean().to_csv('ref_brooklyn.csv')
```

```python
# calculate ref period from 1961-1990 (year)

ref = df.groupby('year')['temperature'].mean().to_frame()
ref['ref'] = ref['temperature'].mean()
ref.to_csv('processed/ref_period_brooklyn.csv')
ref
```
